# Apriori Algorithm 

```python
frequent_itemsets = apriori(movies, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets = apriori(movies, min_support=0.1, use_colnames=True)
frequent_itemsets
```

We will use this min support of 0.1 as 0.2 gives very less combinations

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=5)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:40]
```

These are top 40 rules with lift ratio 10

Lift ratio > 1 implies that the relationship between the antecedent and the consequent is more significant than would be expected if the two sets were independent

# Scatter plots comparing support, confidence and lift

```python
import matplotlib.pyplot as plt
plt.scatter(rules['support'], rules['confidence'],   alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
plt.scatter(rules['support'], rules['lift'],   alpha=0.5)
plt.xlabel('support')
plt.ylabel('lift') 
plt.show()
```

```python
plt.scatter(rules['lift'], rules['confidence'],   alpha=0.5)
plt.xlabel('lift')
plt.ylabel('confidence') 
plt.show()
```
